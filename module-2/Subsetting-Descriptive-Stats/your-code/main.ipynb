####  What is the average purchase price?

```python
data_df['amount_spent'].mean()
```

#### What are the highest and lowest purchase prices? 

```python
print("price max :", data_df['amount_spent'].max(), "price min : ", data_df['amount_spent'].min())
```

#### Select all the customers from Spain.
**Hint**: Remember that you are not asked to find orders from Spain but customers. A customer might have more than one order associated. 

```python
data_df['amount_spent'].mean()
```

#### How many customers do we have in Spain?

```python
data_df.groupby('CustomerID')["Country"].value_counts()
```

```python
data_df_spain = data_df[data_df.Country == 'Spain']

len(data_df_spain['CustomerID'].unique())
```

#### Select all the customers who have bought more than 50 items.
**Hint**: Remember that you are not asked to find orders with more than 50 items but customers who bought more than 50 items. A customer with two orders of 30 items each should appear in the selection.

#### Select orders from Spain that include more than 50 items.

```python
data_df_spain1 = data_df_spain.groupby(['InvoiceNo','CustomerID'])['Quantity'].sum()

total_orders_spain = data_df_spain1[data_df_spain1 > 50]

len(total_orders_spain)

# 29 clients espagnols ont commandé plus de 50 items
```

#### Select all free orders.

```python
data_df[data_df.amount_spent == 0]
```

#### Select all orders whose description starts with `lunch bag`.
**Hint**: use string functions.

```python
#data_lunchbag = [description for description in data_df.Description if str(description).startswith('lunch bag')]

data_lunchbag = data_df.Description.str.startswith('lunch bag')

data_lunchbag_df = data_df[data_lunchbag]

data_lunchbag_df
```

#### Select all `lunch bag` orders made in 2011.

```python
data_lunchbag_df[data_lunchbag_df.year == 2011]
```

#### Show the frequency distribution of the amount spent in Spain.

```python
data_df_spain.groupby(['InvoiceNo'])['amount_spent'].sum().hist(bins = 30)
```

#### Select all orders made in the month of August.

```python
data_august = data_df[data_df.month == 8]
```

#### Find the number of orders made by each country in the month of August.
**Hint**: Use value_counts().

```python
data_per_country_august = data_august.groupby(['InvoiceNo','Country'], as_index=False)['Quantity'].sum()

data_per_country_august.Country.value_counts()
```

#### What's the  average amount of money spent by country?

```python
data_per_country = data_df.groupby(['InvoiceNo','Country'], as_index=False)['amount_spent'].sum()

data_per_country.groupby(['Country'])['amount_spent'].mean()
```
